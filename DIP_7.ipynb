```python
import cv2 as cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pylab as plt
from  PIL import Image
from matplotlib import pyplot as plt
```

#Image  

```python
import cv2

# Load the image
img = cv2.imread("/content/Edge detectn.bmp")

# Resize the image
scale_percent = 50  # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
resized_img = cv2.resize(img, dim, interpolation=cv2.INTER_AREA)
cv2_imshow(resized_img)
```

#  Roberts operator

```python
def roberts(img):
    # Define Roberts kernels
    kernel1 = np.array([[1, 0], [0, -1]], dtype=np.float32)
    kernel2 = np.array([[0, 1], [-1, 0]], dtype=np.float32)

    # Convert image to grayscale
    img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply kernels to image
    gx = cv2.filter2D(img_gray, -1, kernel1)
    gy = cv2.filter2D(img_gray, -1, kernel2)

    # Compute magnitude of gradients
    mag = np.abs(gx) + np.abs(gy)
    mag = np.uint8(mag)

    # Return gradients and magnitude
    return gx, gy, mag

# Load an image
image_path = '/content/Edge detectn.bmp'
image = cv2.imread(image_path)

# Apply Roberts masks
gx, gy, mag = roberts(image)

# Increase the size of the plot
plt.figure(figsize=(15, 5))

# Gradient X
plt.subplot(131)
plt.imshow(gx, cmap='gray')
plt.title('Gradient X')
plt.axis('off')

# Gradient Y
plt.subplot(132)
plt.imshow(gy, cmap='gray')
plt.title('Gradient Y')
plt.axis('off')

# Magnitude
plt.subplot(133)
plt.imshow(mag, cmap='gray')
plt.title('Magnitude')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Prewitt operator

```python
def prewitt(image):
    # Define Prewitt kernels
    kernel_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32)
    kernel_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=np.float32)

    # Convert image to grayscale
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply kernels to image
    x_edges = cv2.filter2D(image_gray, -1, kernel_x)
    y_edges = cv2.filter2D(image_gray, -1, kernel_y)

    # Compute magnitude of gradients
    edges = np.abs(x_edges) + np.abs(y_edges)
    edges = np.uint8(edges)

    return x_edges, y_edges, edges

# Load an image
image_path = '/content/Edge detectn.bmp'
image = cv2.imread(image_path)

# Apply Prewitt masks
x, y, edges = prewitt(image)

# Increase the size of the plot
plt.figure(figsize=(15, 7.5))

# X axis component
plt.subplot(131)
plt.imshow(x, cmap='gray')
plt.title('X-axis Component')
plt.axis('off')

# Y axis component
plt.subplot(132)
plt.imshow(y, cmap='gray')
plt.title('Y-axis Component')
plt.axis('off')

# Aggregate image
plt.subplot(133)
plt.imshow(edges, cmap='gray')
plt.title('Aggregate Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Sobel operator

```python
def sobel(image):
    # Define Sobel kernels
    kernel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
    kernel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32)

    # Convert image to grayscale
    image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply kernels to image
    x_edges = cv2.filter2D(image_gray, -1, kernel_x)
    y_edges = cv2.filter2D(image_gray, -1, kernel_y)

    # Compute magnitude of gradients
    edges = np.abs(x_edges) + np.abs(y_edges)
    edges = np.uint8(edges)

    return x_edges, y_edges, edges

# Load an image
image_path = '/content/Edge detectn.bmp'
image = cv2.imread(image_path)

# Apply Sobel masks
x, y, edges = sobel(image)

# Increase the size of the plot
plt.figure(figsize=(15, 7.5))

# X axis component
plt.subplot(131)
plt.imshow(x, cmap='gray')
plt.title('X-axis Component')
plt.axis('off')

# Y axis component
plt.subplot(132)
plt.imshow(y, cmap='gray')
plt.title('Y-axis Component')
plt.axis('off')

# Aggregate image
plt.subplot(133)
plt.imshow(edges, cmap='gray')
plt.title('Aggregate Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
